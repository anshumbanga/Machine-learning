```python
# import major libraries
import pandas as pd 
import numpy as np 
import matplotlib.pyplot as plt 
```

```python
df = pd.read_csv('placement.csv')
df.head()
```

```python
plt.figure(figsize=(12,4))
plt.scatter(df.cgpa,df.package)
plt.xlabel('CGPA')
plt.ylabel('Package')
plt.show()
```

```python
df.shape
```

```python
x = df.iloc[:,0:1] # dataframe # input 
y = df.package
```

```python
x.values.ndim
```

```python
# importing train test split 

from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2,random_state=101)
```

```python
plt.figure(figsize=(12,4))
plt.scatter(df.cgpa,df.package)
plt.scatter(x_train,y_train)  
plt.xlabel('CGPA')
plt.ylabel('Package')
plt.show()
```

```python
print('rows for training\t',x_train.shape[0])
print('rows for testing\t',x_test.shape[0])
```

```python
from sklearn.linear_model import LinearRegression

lr = LinearRegression()
lr
```

```python
# fitting x_train,y_train

lr.fit(x_train,y_train)
```

```python
# predictions

y_pred = lr.predict(x_test)
```

```python
# y_test-y_pred
```

```python
# r2 score 

from sklearn.metrics import r2_score
```

```python
r2_score(y_test,y_pred)*100
```

```python
plt.figure(figsize=(12,4))
plt.scatter(df.cgpa,df.package,label='Testing',c='#1f77b4')
plt.scatter(x_train,y_train,label='Training',c='#2ca02c')  
plt.plot(x_test,y_pred,c='#e50914',label='Best Fit Line')
plt.xlabel('CGPA')
plt.ylabel('Package')
plt.legend()
plt.show()
```
